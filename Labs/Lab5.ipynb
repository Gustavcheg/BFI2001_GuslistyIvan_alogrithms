### Кривая Гиьтберта

```python
# Кривая Гиьтберта
import turtle

def curveHilbert(lvl, an, s):
    if (lvl == 0):
        t.fd(s)
        return

    t.right(an)
    curveHilbert(lvl - 1, -an, s)

    t.forward(s)
    t.left(an)
    curveHilbert(lvl - 1, an, s)

    t.forward(s)
    curveHilbert(lvl - 1, an, s)

    t.left(an)
    t.forward(s)
    curveHilbert(lvl - 1, -an, s)
    t.right(an)


t = turtle.Turtle()
t.speed('fastest')
t.hideturtle()

level = int(input())
size = 100

curveHilbert(level, 90, size / (2**(level-1)))
turtle.done()
```

### Кривая Серпинского

```python
# Кривая Серпинского
# Использование L-системы

import turtle


def curveSerpinskie(lvl, max_lvl, ls):
    if lvl == max_lvl:
        return ls
    else:
        newls = ''
        for i in ls:
            if i == '-':
                newls += '-'
            elif i == '+':
                newls += '+'
            elif i == 'F':
                newls += 'F'
            elif i == 'X':
                newls += 'XF-F+F-XF+F+XF-F+F-X'

        ls = newls

        return curveSerpinskie(lvl + 1, max_lvl, ls)


t = turtle.Turtle()
t.speed(100000)
t.hideturtle()

max_level = int(input())
sizeF = 20
angle = 90
L_system = "F+XF+F+XF"

# turtle.tracer(0, 0)
curve_Serpinskie_ls = curveSerpinskie(1, max_level, L_system)

for i in curve_Serpinskie_ls:
    if i == '-':
        t.rt(angle)
    elif i == "+":
        t.lt(angle)
    elif i == "F":
        t.fd(sizeF)

turtle.done()
```

### Треугольник \ Салфетка Серпинского

```python
# Треугольник \ салфетка Серпинского

import turtle

def triangleSerpinskie(lvl, s):
    if lvl == 0:
        for _ in range(3):
            t.fd(s)
            t.left(120)
        return

    triangleSerpinskie(lvl - 1, s)
    t.fd(s * 2**(lvl-1))
    
    triangleSerpinskie(lvl - 1, s)
    t.left(120)
    t.fd(s * 2**(lvl-1))
    t.right(120)
    
    triangleSerpinskie(lvl - 1, s)
    t.right(120)
    t.fd(s * 2**(lvl-1))
    t.left(120)

t = turtle.Turtle()
t.speed('fastest')
t.hideturtle()

level = int(input())
size = 200

triangleSerpinskie(level, size / 2**level)

turtle.done()
```

### Ковер Серпинского

### Дерево Пифагора

```python
# Дерево Пифагора
import turtle


def treePythagoras(turt, l, s):
    if (l == 0):
        for _ in range(4):
            turt.fd(s)
            turt.rt(90)
        return

    for _ in range(4):
        turt.fd(s)
        turt.rt(90)

    turt.fd(s)
    another_turt = turt.clone()
    turt.lt(45)
    treePythagoras(turt, l - 1, (s * 2**(1/2)) / 2)

    another_turt.rt(45)
    another_turt.fd((s * 2**(1/2)) / 2)
    treePythagoras(another_turt, l - 1, (s * 2 ** (1 / 2)) / 2)


t = turtle.Turtle()
t.speed('fastest')
t.hideturtle()

level = int(input())
step = 100

# turtle.tracer(0, 0)
t.left(90)
treePythagoras(t, level, step)

turtle.done()
```
