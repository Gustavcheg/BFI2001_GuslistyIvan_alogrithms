Реализовать следующие структуры данных:
  
● <b> Стек (stack): </b>  
Операции для стека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, извлечение элемента из начала;  </i>  

```python
class Stack():
    def __init__(self):
        self.elements = []
        
    def isEmpty(self):
        return self.elements == []
        
    def push(self, el):
        self.elements.insert(0, el)
        
    def pop(self):
        if (self.isEmpty()):
            return None
        else:
            return self.elements.pop(0)
        
    def get(self):
        if (self.isEmpty()):
            return None
        else:
            return self.elements[0]
        
    def printStack(self):
        for el in self.elements:
            print(el, end=' ')
            
        print()
```

● <b> Дек (двусторонняя очередь, deque):  </b>  
Операции для дека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, добавление нового элемента в конец, извлечение элемента из начала, извлечение элемента из конца. </i>  
  
<i> Разработать программу обработки данных, содержащихся в заранее подготовленном txt-файле, в соответствии с заданиями, применив указанную в задании структуру данных. Результат работы программы вывести на экран и сохранить в отдельном txt-файле. </i>

```python
class Deque():
    def __init__(self, arr=[]):
        self.elements = arr
        
    def isEmpty(self):
        return self.elements == []
        
    def pushRight(self, el):
        self.elements.append(el)
        
    def popRight(self):
        if (self.isEmpty()):
            return None
        else:
            return self.elements.pop()
    
    def pushLeft(self, el):
        self.elements.insert(0, el)
        
    def popLeft(self):
        if (self.isEmpty()):
            return None
        else:
            return self.elements.pop(0)
        
    def isEmpty(self):
        return self.elements == []    
    
    def getLeft(self):
        if (self.isEmpty()):
            return None
        else:
            return self.elements[-1]
    
    def getRight(self):
        if (self.isEmpty()):
            return None
        else:
            return self.elements[0]
        
    def printDeq(self):
        for el in self.elements:
            print(el, end=' ')     
        print()
        
```

### Задание №1
Отсортировать строки файла, содержащие названия книг, в алфавитном порядке с использованием двух деков.

```python
filename = "TXTs/ex1.txt"

with open(filename, mode='r', encoding='utf-8') as f:
    deq1 = Deque(arr=f.readlines())
    deq2 = Deque(arr=[])
    
    f_sorted = True
    
    el = deq1.popLeft()
    deq2.pushLeft(el)
    
    while (not deq1.isEmpty()):
        el = deq1.popLeft()
        if (ord(deq2.getLeft()[0]) <= ord(el[0])):
            deq2.pushRight(el)
                
        else:
            deq2.pushLeft(el)
            f_sorted = False
            
        if (f_sorted == False and deq1.isEmpty()):
            deq1, deq2 = deq2, Deque(arr=[])
            el = deq1.popLeft()
            deq2.pushLeft(el)
            f_sorted = True

    deq2.printDeq()
```

### Задание №2
Дек содержит последовательность символов для шифровки сообщений. Дан текстовый файл, содержащий зашифрованное сообщение. Пользуясь деком, расшифровать текст. 
Известно, что при шифровке каждый символ сообщения заменялся следующим за ним в деке по часовой стрелке через один.

```python
filename = "TXTs/ex2.txt"

def decrypt(c_text, keyDeq):
    decr_text = c_text
    
    for i in range(len(decr_text)):
        encrypted = False
        
        while (not encrypted):
            deq_s = keyDeq.popRight()
            keyDeq.pushLeft(deq_s)
            
            if(decr_text[i] == deq_s):
                for j in range(2):
                    temp = keyDeq.popRight()
                    keyDeq.pushLeft(temp)
                    
                decr_text[i] = temp
                encrypted = True
                
    return decr_text

with open(filename, mode='r', encoding='utf-8') as f:
    keyDeq = Deque([])
    
    # Алфафит для шифровки
    for i in range(97, 123):
        keyDeq.pushRight(chr(i))
    
    c_text = list(f.readline())
    
    decr_text = decrypt(c_text, keyDeq)    
    
    print("".join(decr_text))
```

### Задание №3
Даны три стержня и n дисков различного размера. Диски можно надевать на стержни, образуя из них башни. Перенести n дисков со стержня А на стержень С, сохранив их первоначальный порядок. При переносе дисков необходимо соблюдать следующие правила:
- на каждом шаге со стержня на стержень переносить только один диск;
- диск нельзя помещать на диск меньшего размера;
- для промежуточного хранения можно использовать стержень В.
Реализовать алгоритм, используя три стека вместо стержней А, В, С. Информация о дисках хранится в исходном файле.

```python
filename = "TXTs/ex3.txt"

def moveDisk(fromRod, toRod):
    toRod.push(fromRod.pop())

# Ханойская башенка
# Рекурсивно перемещаем диски с 1 стержня на 3 через 2 
def ToH(height, fromRod, midRod, toRod):

    if (height >= 1):
        # Перемещаем все диски, кроме нижнего с 1 на 2 стержень
        ToH(height - 1, fromRod, toRod, midRod)
        moveDisk(fromRod, toRod)
        
        ToH(height - 1, midRod, fromRod, toRod)

with open(filename, mode='r', encoding='utf-8') as f:

    roda = Stack()
    
    disks = f.readlines()
    for i in disks:
        roda.push(int(i))
        
    roda.printStack()
    
    rodb = Stack()
    rodc = Stack()
    
    ToH(len(disks), roda, rodb, rodc)
        
    print("Rod A: ", end=' ')
    roda.printStack()
    
    print("Rod B: ", end=' ')
    rodb.printStack()
    
    print("Rod C: ", end=' ')
    rodc.printStack()
```

### Задание №4
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс круглых скобок в тексте, используя стек.

```python
filename = "TXTs/ex4.txt"

def roundParBalance(pars):
    i = 0
    while (not pars.isEmpty()):
        par = pars.pop()
        
        if (par == ')'):
            i += 1
        else:
            i -= 1
            
    if (i == 0):
        return True
    else:
        return False
        

with open(filename, mode='r', encoding='utf-8') as f:

    pars = Stack()
    
    code = f.read()
    
    for i in code:
        if (i == '(' or i == ')'):
            pars.push(i)
    
    if (roundParBalance(pars)):
        print("Скобки в порядке")
    else:
        print("Нарушена целостность кода")
```

### Задание №5

```python
filename = "TXTs/ex5.txt"

def squareParBalance(pars):
    i = 0
    while (not pars.isEmpty()):
        par = pars.popRight()
        
        if (par == ']'):
            i += 1
        else:
            i -= 1
            
    if (i == 0):
        return True
    else:
        return False
        

with open(filename, mode='r', encoding='utf-8') as f:

    pars = Deque()
    
    code = f.read()
    
    for i in code:
        if (i == '[' or i == ']'):
            pars.pushRight(i)
    
    if (squareParBalance(pars)):
        print("Скобки в порядке")
    else:
        print("Нарушена целостность кода")
```

### Задание №8
Дан текстовый файл. Используя стек, сформировать новый текстовый файл, содержащий строки исходного файла, записанные в обратном порядке: первая строка становится последней, вторая – предпоследней и т.д.

```python
filename = "TXTs/ex8.txt"

with open(filename, mode='r', encoding='utf-8') as f:
    stc = Stack()
    
    strings = f.readlines()

    for i in strings:
        stc.push(i)

    filename_out = "TXTs/ex8_out.txt"
    
    with open(filename_out, mode='w', encoding='utf-8') as f_out:
        while (not stc.isEmpty()):
            f_out.write(stc.pop())
```
