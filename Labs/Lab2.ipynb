```python
# Бинарный поиск
# Итерационный метод
def binarySearchIter(arr, el, low, high):
    while low <= high:
            
        mid = (low + high) // 2
        if (arr[mid] == el):
            return mid
        elif (arr[mid] < el):
            low = mid + 1
        else:
            high = mid - 1
        
    return -1

# Рекурсивный метод
def binarySearchRecur(arr, el, low, high):
    if (low > high):
        return -1
        
    mid = (low + high) // 2

    if (arr[mid] == el):
        return mid
    elif (arr[mid] < el):
        return binarySearchRecur(arr, el, mid + 1, high)
    else:
        return binarySearchRecur(arr, el, low, mid - 1)
        
        
user_el = int(input("Введите элемент, который нужно найти: "))
print("Элемент: ", user_el) 

start_time = time.time()
found_ind = binarySearchIter(rand_arr, user_el, 0, len(rand_arr) - 1)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print("Элемент {} находится на позиции {}".format(user_el, found_ind))
```

```python
#Метод Фибоначчи
# Разделя и властвуй

# Пользовательская функция нахождения минимума из двух
def minEl(a, b):
    return ((a + b) - abs(a - b)) // 2

# Поиск использует ряд Фибоначчи
def FibonacciSearch(arr, el):
    # Используем три соседних числа из ряда
    fib_prev_prev = 0
    fib_prev = 1
    fib_cur = fib_prev + fib_prev_prev
    
    # Находим ближайшее к длине массива число Фибоначчи 
    while (fib_cur < len(arr)):
        fib_prev_prev = fib_prev
        fib_prev = fib_cur
        fib_cur = fib_prev + fib_prev_prev
        
    index = -1
    while (fib_cur > 1):
        # Выбираем наименьший элемент, или длина массива или первое число ряда
        i = minEl(index + fib_prev_prev, len(arr) - 1)
        
        # Если текущий элемент меньше искомого, то перемещаем числа Фибоначчи на один шаг вниз
        if (arr[i] < el):
            fib_cur = fib_prev
            fib_prev = fib_prev_prev
            fib_prev_prev = fib_cur - fib_prev
            index = i
            
        # Если текущий элемент больше искомого, то перемещаем числа Фибоначчи на один два шага вниз
        elif (arr[i] > el):
            fib_cur = fib_prev_prev
            fib_prev = fib_prev - fib_prev_prev
            fib_prev_prev = fib_cur - fib_prev
            
        # Еслемент найден
        else:
            return i
        
    if (fib_prev and index < (len(arr) - 1) and arr[index + 1] == el):
        return index + 1
    return -1
    
print(rand_arr)
user_el = int(input("Введите элемент, который нужно найти: "))

print("Элемент: ", user_el)

start_time = time.time()
found_ind = FibonacciSearch(rand_arr, user_el)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print("Элемент {} находится на позиции {}".format(user_el, found_ind)) 
```

Интерполяционный поиск

```python
#Интерполяционный
# Разделяй и властвуй

# Аналог бинарного поиска, но поиск начинается не всегда с середины
# Поиск вычисляет вероятную позицию искомого элемента по формуле описанной в листинге

def interSearch(arr, el, low, high):

    while low <= high and el >= arr[low] and el <= arr[high]:
        # Вычисляем вероятный индекс 
        # Наиболее низкое, если ближе к arr[low] и наиболее высокое, если ближе к arr[high]
        i = low + int((float(high - low) / (arr[high] - arr[low])) * (el - arr[low]))
        
        if (el == arr[i]):
            return i
        elif (el < arr[i]):
            low = i + 1
        else:
            high = i - 1
    return -1 

print(rand_arr)
user_el = int(input("Введите элемент, который нужно найти: "))

print("Элемент: ", user_el)

start_time = time.time()
found_ind = interSearch(rand_arr, user_el, 0, len(rand_arr) - 1)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print("Элемент {} находится на позиции {}".format(user_el, found_ind)) 
```

### Задание №3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.


```python
import random

# Функция которая проверяет конфликn
# Не может быть более одного ферзя в столбце, в строке и по диагоналям

# Используется кортеж позиций ферзей 
# Элемент в кортеже - номер позиции по строке
# Индекс элемента в кортеже - номер позиции по столбцу
def isConflict(state,col):
    # Получаем кол-во неконфликтных строк
    row = len(state)
    
    for i in range(row):
        # Если столбекц ферзя ряда + 1 совпдает с столбцом другого ферзя - конфликт
        if abs(state[i] - col) in (0, row - i): 
            return True
        
    return False

# Ищем все возможные варианты расположения
# В функции используется ключевое слово yield для создания генератора
def queens(num=8,state=()):
    # Генерируем кортеж позиций
    for pos in range(num):
        if not isConflict(state, pos):
            if len(state) == num - 1:
                yield(pos,)
            else:
                # Формируем кортеж позиций
                for result in queens(num, state + (pos,)):
                    yield (pos,) + result
                    
# Печатаем шах доску с позициями                
def printBoardQueens(solution):
    def line(pos, length=len(solution)):
        return '* ' * (pos) + '@ ' + '* ' * (length - pos - 1)
    
    for pos in solution:
        print(line(pos))
    
    return "Одно из расположений: " + str(solution)
        
# for solution in list(queens(8)):
#     print(solution)
        
print('Общее кол-во расположений = ' + str(len(list(queens()))))

choosen = random.choice(list(queens()))
print(printBoardQueens(choosen))
```
