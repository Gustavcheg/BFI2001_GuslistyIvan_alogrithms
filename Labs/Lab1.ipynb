---
> * Сортировка выбором. <br>
 Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
 после чего срез уменьшается и цикл повторяется.

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
#TODO
def selectSort(arr):              
    for el in arr:
        for i in range(len(el)):
            for j in range(i, len(el)):
                if el[i] > el[j]:
                    el[i], el[j] = el[j], el[i]
                    
arr_to_sort1 = copy.deepcopy(rand_arr)

selectSort(arr_to_sort1)
print("[{}, ..., {}]".format(arr_to_sort1[0], arr_to_sort1[-1]))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

---
> * Сортировка вставкой. <br>
 Массив условно разделён на отсорт. и неотсорт. части <br> 
Для текущего элемента нужно найти место в отсортированной части массива <br>
Далее нужно сдвинуть массива вправо (сдвинуть с найденного места) <br>
Вставить элемент на нужное место <br>

```python
# Сортировка вставкой.
start_time = time.time()
#TODO
def insertSort(arr):
    for el in arr:
        for i in range(len(arr)):
            temp = el[i] # Текущий элемент
            j = i - 1
            while j >= 0 and temp < el[j]: # Сдвигаем массив
                el[j + 1] = el[j]
                j -= 1
            el[j + 1] = temp # Вставляем элемент

arr_to_sort2 = copy.deepcopy(rand_arr)

insertSort(arr_to_sort2)
print("[{}, ..., {}]".format(arr_to_sort2[0], arr_to_sort2[-1]))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

---
> * Сортировка обменом. || Сортировка пузырьком. <br>
 Попарно сравниваем элементы, если текущий больше следующего, то меняем местами

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()
#TODO
def exchangeSort(arr):
    for el in arr:
        for i in range(len(arr)):
            for j in range(len(arr)-i-1):
                if el[j] > el[j + 1]:
                    el[j], el[j + 1] = el[j + 1], el[j]

arr_to_sort3 = copy.deepcopy(rand_arr)

exchangeSort(arr_to_sort3)
print("[{}, ..., {}]".format(arr_to_sort3[0], arr_to_sort3[-1]))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

---
> * Сортировка Шелла. <br>
Better inserts, улучшение достигается путём сравнение элементов на некотором расстоянии, которое
со временем выполнения алгоритма уменьшается, пока массив не будет отсортирован

```python
# Сортировка Шелла.
start_time = time.time()
#TODO
def shellSort(arr):
    
    for el in arr:
        inc = len(arr) // 2
        while inc:
            for i in range(inc, len(el)):
                temp = el[i]
                j = i

                while j >= inc and el[j - inc] > temp:
                    el[j] = el[j - inc]
                    j -= inc

                el[j] = temp

            inc //= 2

arr_to_sort4 = copy.deepcopy(rand_arr)

shellSort(arr_to_sort4)
print("[{}, ..., {}]".format(arr_to_sort4[0], arr_to_sort4[-1]))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

> * Быстрая сортировка. <br>
 Сортировка рекурсивная, для начала нужно определеить опорный элемент, начальный и конечный (их индексы) <br>
 Массив разделяется по опорному элементу, влево - элементы меньшие, вправо - большие <br>
 Этот процесс повторяется для массива слева от опорного, 
 а также для массива элементов справа, пока весь массив не будет отсортирован.

```python
# Быстрая сортировка.
start_time = time.time()
#TODO
def quickSort(arr, start, end):
    if start >= end:
        return
    i, j = start, end
    
    pivot = arr[random.randint(start, end)]
    
    while i <= j:
        while arr[i] < pivot:
            i += 1
            
        while arr[j] > pivot:
            j -= 1
            
        if i <= j:
            arr[i], arr[j] = arr[j], arr[i]
            i, j = i + 1, j - 1
            
    quickSort(arr, start, j)
    quickSort(arr, i, end)
    
def startQuickSort(arr):
    for el in arr:
        quickSort(el, 0, len(el) - 1)


arr_to_sort6 = copy.deepcopy(rand_arr)

startQuickSort(arr_to_sort6)
print("[{}, ..., {}]".format(arr_to_sort6[0], arr_to_sort6[-1]))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))  
```

---
> * Турнирная сортировка.
> В данном алгоритме сортировки была использована куча, как приоритетная очередь
В качестве реализации кучи был выбран массив. Метод heapifyMy превращает массив в кучу или в приоритетную очередь. <br>
Полученная куча - отдельный массив, то есть турнирная сетка (размером 7 элементов), в которой путём кучевания (heapifyMy) на верх (в начало) перемещается наименьший элемент (победитель) <br>
В ходе сортировки могут быть выявлены элементы, которые меньше последнего победителя, тогда он становится проигравшим и после сортировки, если есть проигравшие, сортировка повторяется

```python
# Турнирная сортировка.
start_time = time.time()
#TODO

# Проверяем массив на длину, если меньше 7, то кучей будет весь массив
def checkArr(arr):
    if (len(arr) >= 7):
        return [7, arr[:7]]
    else:
        return [0, arr[:]]

# Кучевание 
def heapifyMy(arr):
    n = len(arr)
    for i in range(n - 1, -1, -1):
        largest = i

        left = 2*i + 1
        right = 2*i + 2

        if (right < n and arr[right] < arr[largest]):
            largest = right

        if (left < n and arr[left] < arr[largest]):
            largest = left

        if (largest != i):
            arr[i], arr[largest] = arr[largest], arr[i]

            heapifyMy(arr)

# Турнирная сортировка
def winnerWinnerChickenDinnerSort(arr): 
#     Проверяем массив
    # arr_len, tourHeap = checkArr(arr)[0], checkArr(arr)[1]
    if(len(arr) >= 7):
        arr_len = 7
        tourHeap = arr[:7]
    else:
        arr_len = 0
        tourHeap = arr[:]
    heapifyMy(tourHeap)
    
#     Победители, проигравшие
    winners = [] 
    losers = []
    
#     Определяем статус элемента (победитель, проигравший)
    for i in range(arr_len, len(arr)):
        cur_element = tourHeap.pop(0)
        
#         Втсавляем первый элемент в победителя
        if (i == arr_len):
            winners.append(cur_element)
        elif (cur_element >= winners[-1]):
            winners.append(cur_element)
        else:
            losers.append(cur_element)

#         Нужно только, когда длина кучи = длине массива, чтобы не выйти за границы массива
        if(arr_len >= 7):
            tourHeap.append(arr[i])
            
        heapifyMy(tourHeap)
        
#     Очищаем кучу
    for i in range(arr_len):
        cur_element = tourHeap.pop(0)
        
        if (cur_element >= winners[-1]):
            winners.append(cur_element)
        else:
            losers.append(cur_element)
        
        heapifyMy(tourHeap)
    
#     Если есть проигравшие - повторяем 
    if(losers):
        return winnerWinnerChickenDinnerSort(losers + winners)
    else:
        return winners
    
    
def tournamentSort(arr):
    for i in range(len(arr)):
        arr[i] = winnerWinnerChickenDinnerSort(arr[i])
        

arr_to_sort5 = copy.deepcopy(rand_arr)

tournamentSort(arr_to_sort5)
print("[{}, ..., {}]".format(arr_to_sort5[0], arr_to_sort5[-1]))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
