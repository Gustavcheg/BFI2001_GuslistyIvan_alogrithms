#### Упрощенный алгоритм Бойера-Мура

```python
# Функция формирования таблицы смещений
def predFunc(substr):
    d = {}
    temp = set()
    m = len(substr)
    
    for i in range(m-2, -1, -1):
        # Не был ли символ уже в строке
        if (substr[i] not in temp):
            d[substr[i]] = m - i - 1
            temp.add(substr[i])
    
    if substr[-1] not  in temp:
        d[substr[-1]] = m
        
    d['*'] = m
    
    return d

def BMSearch(string, substr):
    n = len(string)
    m = len(substr)
    
    # Получаем смещения
    offsets = predFunc(substr)
    
    if (n >= m):
        # Задаём счётчик проверяемого символа
        i = m - 1
        
        while (i < n):
            # Для проверик символа смещения
            k = 0
            for j in range(m - 1, -1, -1):
                if (string[i - k] != substr[j]):
                    # Определяем смещение для последнего символа
                    if (j == m - 1):
                        offset = offsets[string[i]] if offsets.get(string[i], False) else offsets['*']
                    else:
                    # Если всё таки последний символ совпал, то смещение для следующих символов
                        offset = offsets[substr[j]]
                    
                    # Переходим дальеш по смещению
                    i += offset
                    break
                    
                k += 1
            
            # Дошли до конца, соответственно подстрока найдена
            if (j == 0):
                return i - k +1
            
    return 0


mystr = input("Введите строку: ")
substring = input("Введите подстроку: ")

print(mystr)
print(substring)

start_time = t.now()
findSub = BMSearch(mystr, substring)
print("--- {0} ms ---".format((t.now() - start_time)))

if (findSub):
    print("Индекс 1-го вхождения в подстроку: ", findSub)
else:
    print("Не найдено!")
```
