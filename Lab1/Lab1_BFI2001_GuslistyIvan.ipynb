### Задание 3.1 - Сортировка выбором (по возрастанию)
> Сортируем строки в матрице. <br>
C помощью двух циклов ищем (выбираем) наибольший элемент из неостортированной части массива 
и меняем местами с текущим

```python
def selectSort(arr):              
    for el in arr:
        for i in range(m):
            for j in range(i, m):
                if el[i] > el[j]:
                    el[i], el[j] = el[j], el[i]
    #return arr
```

### Задание 3.2 - Сортировка вставкой (по возрастанию)
> Массив условно разделён на отсорт. и неотсорт. части <br> 
Для текущего элемента нужно найти место в отсортированной части массива <br>
Далее нужно сдвинуть массива вправо (сдвинуть с найденного места) <br>
Вставить элемент на нужное место <br>

```python
def insertSort(arr):
    for el in arr:
        for i in range(m):
            temp = el[i] # Текущий элемент
            j = i - 1
            while j >= 0 and temp < el[j]: # Сдвигаем массив
                el[j + 1] = el[j]
                j -= 1
            el[j + 1] = temp # Вставляем элемент
    #return arr   
```

> Замеряем время

```python
import copy
arr_to_sort2 = copy.deepcopy(rand_arr)

%time insertSort(arr_to_sort2)
print("[{}, ..., {}]".format(arr_to_sort2[0], arr_to_sort2[-1]))
```

### Задание 3.3 - Сортировка обменом (по возрастанию) или пузырьковая сортировка
> Попарно сравниваем элементы, если текущий больше следующего, то меняем местами

```python
def exchangeSort(arr):
    for el in arr:
        for i in range(m):
            for j in range(m-i-1):
                if el[j] > el[j + 1]:
                    el[j], el[j + 1] = el[j + 1], el[j]
    # return arr
```

### Задание 3.4 - Сортировка Шелла (по возрастанию)
> Better inserts, улучшение достигается путём сравнение элементов на некотором расстоянии, которое
со временем выполнения алгоритма уменьшается, пока массив не будет отсортирован

```python
def shellSort(arr):
    
    for el in arr:
        inc = n // 2
        while inc:
            for i in range(inc, len(el)):
                temp = el[i]
                j = i

                while j >= inc and el[j - inc] > temp:
                    el[j] = el[j - inc]
                    j -= inc

                el[j] = temp

            inc //= 2

    #return arr
```

> Замеряем время

```python
import copy
arr_to_sort4 = copy.deepcopy(rand_arr)

%time shellSort(arr_to_sort4)
print("[{}, ..., {}]".format(arr_to_sort4[0], arr_to_sort4[-1]))
```

### Задание 3.5 - Турнирная сортировка (по возрастанию)
> В данном алгоритме сортировки была использована куча, как приоритетная очередь
В качестве реализации кучи был выбран массив. Метод heapifyMy превращает массив в кучу или в приоритетную очередь. <br>
Полученная куча - отдельный массив, то есть турнирная сетка (размером 7 элементов), в которой путём кучевания (heapifyMy) на верх (в начало) перемещается наименьший элемент (победитель) <br>
В ходе сортировки могут быть выявлены элементы, которые меньше последнего победителя, тогда он становится проигравшим и после сортировки, если есть проигравшие, сортировка повторяется

```python
# Проверяем массив на длину, если меньше 7, то кучей будет весь массив
def checkArr(arr):
    if (len(arr) >= 7):
        return [7, arr[:7]]
    else:
        return [0, arr[:]]

# Кучевание 
def heapifyMy(arr):
    n = len(arr)
    for i in range(n - 1, -1, -1):
        largest = i

        left = 2*i + 1
        right = 2*i + 2

        if (right < n and arr[right] < arr[largest]):
            largest = right


        if (left < n and arr[left] < arr[largest]):
            largest = left

        if (largest != i):
            arr[i], arr[largest] = arr[largest], arr[i]

            heapifyMy(arr)

# Турнирная сортировка
def winnerWinnerChickenDinnerSort(arr): 
#     Проверяем массив
    arr_len, tourHeap = checkArr(arr)[0], checkArr(arr)[1]
    heapifyMy(tourHeap)
    
#     Победители, проигравшие
    winners = [] 
    losers = []
    
#     Определяем статус элемента (победитель, проигравший)
    for i in range(arr_len, len(arr)):
        cur_element = tourHeap.pop(0)
        
#         Втсавляем первый элемент в победителя
        if (i == arr_len):
            winners.append(cur_element)
        elif (cur_element >= winners[-1]):
            winners.append(cur_element)
        else:
            losers.append(cur_element)

#         Нужно только, когда длина кучи = длине массива, чтобы не выйти за границы массива
        if(arr_len >= 7):
            tourHeap.append(arr[i])
            
        heapifyMy(tourHeap)
        
#     Очищаем кучу
    for i in range(arr_len):
        cur_element = tourHeap.pop(0)
        
        if (cur_element >= winners[-1]):
            winners.append(cur_element)
        else:
            losers.append(cur_element)
        
        heapifyMy(tourHeap)
    
#     Если есть проигравшие - повторяем 
    if(losers):
        return winnerWinnerChickenDinnerSort(losers + winners)
    else:
        return winners
    
    
def tournamentSort(arr):
    for i in range(len(arr)):
        arr[i] = winnerWinnerChickenDinnerSort(arr[i])
```

> Замеряем время

```python
import copy
arr_to_sort5 = copy.deepcopy(rand_arr)

%time tournamentSort(arr_to_sort5)
print("[{}, ..., {}]".format(arr_to_sort5[0], arr_to_sort5[-1]))
```

### Задание 3.6 - Быстрая сортировка (по возрастанию)
> Сортировка рекурсивная, для начала нужно определеить опорный элемент, начальный и конечный (их индексы) <br>
 Массив разделяется по опорному элементу, влево - элементы меньшие, вправо - большие <br>
 Этот процесс повторяется для массива слева от опорного, 
 а также для массива элементов справа, пока весь массив не будет отсортирован.

```python
import random

def quickSort(arr, start, end):
    if start >= end:
        return
    i, j = start, end
    
    pivot = arr[random.randint(start, end)]
    
    while i <= j:
        while arr[i] < pivot:
            i += 1
            
        while arr[j] > pivot:
            j -= 1
            
        if i <= j:
            arr[i], arr[j] = arr[j], arr[i]
            i, j = i + 1, j - 1
            
    quickSort(arr, start, j)
    quickSort(arr, i, end)
        
    
    #return arr
    
def startQuickSort(arr):
    for el in arr:
        quickSort(el, 0, len(el) - 1)
```

> Замеряем время

```python
import copy
arr_to_sort6 = copy.deepcopy(rand_arr)

%time startQuickSort(arr_to_sort6)
print("[{}, ..., {}]".format(arr_to_sort6[0], arr_to_sort6[-1]))
```

### Задание 3.7 - пирамидальная сортировка (по возрастанию)
> Замеряем время
> В качестве реализации кучи был выбран массив
 > * Строим max-heap - куча, где значение в родительском узле больше, чем в дочерних
 > * В корне хранится наибольший элемент, меняем его местами с последним, и уменьшаем размер кучи на 1
 > * Повторяем пока размер кучи больше 1

```python
# Преобразование массива в кучу, только если дочерние узлы уже преобразованы
def heapify(arr, n, i):
    # Родительский узел 
    largest = i
    
    # Вычисляем левый дочерний узел 
    left = 2*i + 1
    # Правый дочерний узел
    right = 2*i + 2
    
    # Проверяем, не вышли ли за границы
    # Если правый больше корня, то перемещаем вверх кучи
    if (right < n and arr[right] > arr[largest]):
        largest = right
    
    # Если левый больше корня, то перемещаем вверх кучи
    if (left < n and arr[left] > arr[largest]):
        largest = left
        
    # Если наибольший элемент - не корень
    if (largest != i):
        arr[i], arr[largest] = arr[largest], arr[i]
        
        heapify(arr, n, largest)

def makeHeapFromArray(arr):
    # Строим нашу кучу из массива
    for i in range(len(arr) - 1, -1, -1):
        # Передаём индекс, чтобы выстроить дочерние элементы
        heapify(arr, len(arr), i)

def heapSort(arr):

    for el in arr:
        makeHeapFromArray(el)
        # Magic. По очереди извлекаем элементы из кучи
        for i in range(len(el) - 1, -1, -1):
            # Меняем местами текущий корень с концом
            el[0], el[i] = el[i], el[0]
        
            heapify(el, i, 0)
    
    #return arr
```

> Замеряем время

```python
import copy
arr_to_sort7 = copy.deepcopy(rand_arr)

%time heapSort(arr_to_sort7)
print("[{}, ..., {}]".format(arr_to_sort7[0], arr_to_sort7[-1]))
```
