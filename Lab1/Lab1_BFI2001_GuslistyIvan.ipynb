```python
%%time
# Задание 2 - Создание матрицы и заполнение случаныйми числами
import random
m = 2
n = 3
min_limit = -250
max_limit = 1000 + 6

rand_arr = [[0] * m for i in range(n)]


for i in range(n):
    for j in range(m):
        rand_arr[i][j] = random.randint(min_limit, max_limit)

print(rand_arr)
```

```python
%%time 
#  Задание 3.1 - Сортировка выбором (по возрастанию)
# Сортируем строки, в матрице отсортированы строки
# С помощью двух циклов ищем (выбираем) наибольший элемент из неостортированной частит массива и меняем местами
# с текущим
arr_to_sort1 = rand_arr.copy()

def selectSort(arr):              
    for el in arr:
        for i in range(m):
            for j in range(i, m):
                if el[i] > el[j]:
                    el[i], el[j] = el[j], el[i]
    return arr

print(selectSort(arr_to_sort1))
```

```python
%%time
# Задание 3.2 - Сортировка вставкой (по возрастанию)
# Массив условно разделён на отсорт. и неотсорт. части 
# Для текущего элемента нужно найти место в отсортированной части массива
# Далее нужно сдвинуть массива вправо (сдвинуть с найденного места)
# Вставить элемент на нужное место

arr_to_sort2 = rand_arr.copy()

def insertSort(arr):
    for el in arr:
        for i in range(m):
            temp = el[i] # Текущий элемент
            j = i - 1
            while j >= 0 and temp < el[j]: # Сдвигаем массив
                el[j + 1] = el[j]
                j -= 1
            el[j + 1] = temp # Вставляем элемент
    return arr

print(insertSort(arr_to_sort2))
            
    
```

```python
%%time
# Задание 3.3 - Сортировка обменом (по возрастанию) или пузырьковая сортировка
# Попарно сравниваем элементы, если текущий больше следующего, то меняем местами
arr_to_sort3 = rand_arr.copy()

def exchangeSort(arr):
    for el in arr:
        for i in range(m):
            for j in range(m-i-1):
                if el[j] > el[j + 1]:
                    el[j], el[j + 1] = el[j + 1], el[j]
    return arr

print(exchangeSort(arr_to_sort3))
```

```python
%%time
# Задание 3.4 - Сортировка Шелла (по возрастанию)
# Better inserts, улучшение достигается путём сравнение элементов на некотором расстоянии, которое
# со временем выполнения алгоритма уменьшается 

arr_to_sort4 = rand_arr.copy()

def shellSort(arr):
    inc = n // 2
    
    for el in arr:
        while inc:
            for i in range(inc, m):
                temp = el[i]
                j = i

                while j >= inc and el[j - inc] > temp:
                    el[j] = el[j - inc]
                    j -= inc

                el[j] = temp

            inc //= 2

    return arr

print(shellSort(arr_to_sort4))
```

```python
%%time
# Задание 3.6 - Быстрая сортировка (по возрастанию)
# Сортировка рекурсивная, для начала нужно определеить опорный элемент, начальный и конечный (их индексы)
# Массив разделяется по опорному элементу, влево - элементы меньшие, вправо - большие 

# Этот процесс повторяется для коллекции слева от опорного, 
# а также для массива элементов справа, пока весь массив не будет отсортирован.


arr_to_sort6 = rand_arr.copy()

def makeParts(arr, start, end):
        pivot = arr[start]
        i = start + 1
        j = end - 1
        
        while True:
            while (i <= j and arr[i] <= pivot):
                i += 1
            while (i <= j and arr[j] >= pivot):
                j -= 1
                
            if i <= j:
                arr[i], arr[j] = arr[j], arr[i]
            else:
                arr[start], arr[j] = arr[j], arr[start]
                return j

def quickSort(arr, start, end):
#     Сортируем каждую строку матрицы
    for el in arr:
        if end - start > 1:
            pivot_ind = makeParts(el, start, end)
            quickSort(el, start, pivot_ind)
            quickSort(el, pivot_ind + 1, end)
    
    return arr

print(quickSort(arr_to_sort6, 0, m))
```
