```python
def find_eq_points(n):
    c_matches = n * (n - 1) // 2
    # ะะฐััะธะฒ ะดะปั ะดะตะผะพะฝัััะฐัะธะธ ะฟัะฐะฒะธะปัะฝะพััะธ ัะฐัะฟัะตะดะตะปะตะฝะธั ะพัะบะพะฒ
    teams = [0] * n
    
    # ะััั ะปะธ ะฒะพะทะผะพะถะฝะพััั ััะณัะฐัั ะฑะตะท ะฝะธััะธั
    if c_matches % n == 0:
        # ะะพะป-ะฒะพ ัะพัะตะดะตะน ะบะพัะพััะต ะดะพะปะถะฝั ะฟัะพะธะณัะฐัั ัะตะบััะตะน ะบะพะผะฐะฝะดะต
        lose_neighbours = c_matches // n
        
        for i in range(n - 1):
            # ะัะพะฒะตัะบะฐ ะบะพะฝัะฐ ะผะฐััะธะฒะฐ, ััะพะฑั ััะตะท ะดะพ ะบะพะฝัะฐ ะฝะต ะฑัะป ะฑะพะปััะต lose_neighbours
            if n - i > lose_neighbours:
                for j in range(lose_neighbours):
                    teams[i] += 3
                    print(1, end=" ")
            else:
                for j in range(n - i - 1):
                    teams[i] += 3
                    print(1, end=" ")
                
            for j in range(i + lose_neighbours + 1, n):
                teams[j] += 3
                print(-1, end=" ")
    
    # ะัะปะธ ะฝะตั ะฒะพะทะผะพะถะฝะพััะธ ััะณัะฐัั ะฑะตะท ะฝะธััะธั, ัะพ ะบะฐะถะดะฐั ะบะพะผะฐะฝะดะฐ ะดะพะปะถะฝะฐ ััะณัะฐัั ะฒ ะฝะธััั ัะพะฒะฝะพ ะพะดะธะฝ ัะฐะท
    else:
        lose_neighbours = c_matches // n
        
        for i in range(n - 1):
            if n - i - 1 > lose_neighbours:
                for j in range(lose_neighbours):
                    teams[i] += 3
                    print(1, end=" ")
                    
                teams[i] += 1
                teams[i + lose_neighbours + 1] += 1
                print(0, end=" ")
            else:
                for j in range(n - i - 1):
                    teams[i] += 3
                    print(1, end=" ")
            
            for j in range(i + lose_neighbours + 2, n):
                teams[j] += 3
                print(-1, end=" ")
              

t = int(input("ะะฒะตะดะธัะต ะบะพะป-ะฒะพ ัะตะทะพะฝะพะฒ: "))

for i in range(t):
    n = int(input("ะะฒะตะดะธัะต ะบะพะป-ะฒะพ ะบะพะผะฐะฝะด ะฒ {} ัะตะทะพะฝะต: ".format(i + 1)))
    find_eq_points(n)
```

```python
### ะะตะฑะพะปััะพะต ััะพัะฝะตะฝะธะต: ัะตัะฝัะต ะฟะพะทะธัะธะธ 0, 2, 4... (ั. ะต. ะฟะตัะฒะฐั (ะฝัะปะธะฒะฐั) ะฟะพะทะธัะธั - ัะตัะฝะฐั)
# ะฝะตัััะฝัะต ะฟะพะทะธัะธะธ: 1, 3, 5...

# ะััะพะดั ะธะท ััะปะพะฒะธั ะผะพะถะฝะพ ัะดะตะปะฐัั ะฒัะฒะพะด, ััะพ length1 + length3 + ... = n = length2 + length4 + ...
# ะะฐะผ ะฝัะถะฝะพ ะผะธะฝะธะผะธะทะธัะพะฒะฐัั ััะฝะบัะธั ั1 * length1 + ั2 * length ...

# ะะฐะทะดะตะปะธะผ ะผะฐััะธะฒ ะฝะฐ ะดะฒะต ัะฐััะธ: ัะปะตะผะตะฝัั ั ัะตัะฝัะผะธ ะฟะพะทะธัะธัะผะธ ะธ ัะปะตะผะตะฝัั ั ะฝะตัััะฝัะผะธ ะฟะพะทะธัะธัะผะธ
# ะะฐะปะตะต ะฝัะถะฝะพ ะฝะฐะนัะธ ะฒ ะพะฑะพะธั ัะฐัััั ะผะธะฝะธะผะฐะปัะฝัะต ัะปะตะผะตะฝัั
# ะัะตะผ ัะปะตะผะตะฝัะฐะผ (ัะตะฝะฐะผ) ะดะพ ะฝะฐะนะดัะฝะฝัั ะผะธะฝะธะผะฐะปัะฝัั ะผะพะถะฝะพ ะทะฐะดะฐัั ะดะปะธะฝั 1, ะฐ ะผะธะฝะธะผะฐะปัะฝัะผ ัะตะฝัะผ ะผะพะถะฝะพ ะทะฐะดะฐัั ะพััะฐะฒััััั ะดะปะธะฝั 
# ะขะพ ะตััั n - (ะบะพะป-ะฒะพ ัะปะตะผะตะฝัะพะฒ ะดะพ ะผะธะฝะธะผะฐะปัะฝะพะณะพ)

# ะะฐะปะตะต ัะปะพะถะธะผ ะฒัะต ะฟะพะปััะตะฝะฝัะต ััะพะธะผะพััะธ ะฟััะตะน ะธ ะฟะพะปััะธะผ ัะตะทัะปััะฐั
def min_path(n, c):
    # ะะธะฝะธะผะฐะปัะฝัะต ัะปะตะผะตะฝัั ะฝะฐ ัััะฝะพะน ะธ ะฝะตัััะฝะพะน ะฟะพะทะธัะธัั
    min_even = 10 ** 9
    min_odd = 10 ** 9
    
    # ะั ะธะฝะดะตะบัั ะฒ ะผะฐััะธะฒะต c[]
    min_even_ind = 0
    min_odd_ind = 0
    
    # ะกัะผะผะฐ ะดะพ ะฝะฐะนะดะตะฝะฝัั ะผะธะฝะธะผะฐะปัะฝัั ัะปะตะผะตะฝัะพะฒ
    s_even = 0
    s_odd = 0

    # ะะพะป-ะฒะพ ะดะพ ะฝะฐะนะดะตะฝะฝัั ะผะธะฝะธะผะฐะปัะฝัั ัะปะตะผะตะฝัะพะฒ
    c_even = 0
    c_odd = 0

    # ะะฐัะพะดะธะผ ะฒัั ะฒััะตะฟะตัะตัะธัะปะตะฝะฝะพะต
    for i in range(n):
        if i % 2 == 0:
            if min_even > c[i]:
                min_even_ind = i
                min_even = c[i]

        else:
            if min_odd > c[i]:
                min_odd_ind = i
                min_odd = c[i]

    # ะกัะธัะฐะตะผ ััะผะผั ะดะพ ะผะธะฝะธะผะฐะปัะฝัั ัะปะตะผะตะฝัะพะฒ
    for i in range(0, min_even_ind, 2):
        s_even += c[i]

    for i in range(1, min_odd_ind, 2):
        s_odd += c[i]

    # ะะพะป-ะฒะพ (ะผะพะถะฝะพ ัะฑัะฐัั, ะฝะพ ัะฐะบ ัะธัะฐะฑะตะปัะฝะตะน)
    c_even = min_even_ind // 2
    c_odd = min_odd_ind // 2

    # ะัะพะณะพะฒะฐั ััะพะธะผะพััั ะบัะฐััะฐะนัะตะณะพ ะฟััะธ
    return s_odd + min_odd * (n - c_odd) + s_even + min_even * (n - c_even)

t = int(input())
res = []

for i in range(t):    
    in_n = int(input())
    in_c = list(map(int, input().split()))
    res.append(min_path(in_n, in_c))
    
print(*res, sep="\n")
```

```python
# ะัะตะดะฟะพะปะฐะณะตะผ ััะพ ะฒัะต ะปะฐะฒัั ะดะพััะฐัััั A
# ะัะดะตะผ ะพัะฝะธะผะฐัั ั ะฝะตะณะพ ัะนัะฐ (ะดะตะฝัะณะธ ะทะฐ ะฟะพะบัะฐัะบั) ะธ ะฟะปะฐัะธัั ะทะฐะฟัะพัะตะฝะฝัะต ะดะปั G
# ะฟะพะบะฐ ัะฐะทะฝะธัะฐ ะผะตะถะดั ะฒัะตะผะธ ะฟะพะปััะธะฝะฝัะผะธ ะธะผะธ ะดะตะฝัะณะฐะผะธ ะฝะต ััะฐะฝะตั <= 500
def uncle_J_task(n, a_eggs_costs, g_eggs_costs):
    # ะะพะฟัััะธะผ ะฒัะต ัะนัะฐ ะฟะพะปััะฐะตั A
    result = ["A"] * n
    # ะัะต ะดะตะฝัะณะธ ะบะพัะพััะต ะพะฝ ะฟะพะปััะธั
    a_total = sum(a_eggs_costs)
    g_total = 0
    i = 0

    # ะะพะบะฐ ัะฐะทะฝะธัะฐ ะผะตะถะดั A ะธ G ะฝะต ััะฐะฝะตั ะผะตะฝััะต 500
    # ะะปะธ ะฟะพะบะฐ ะฝะต ะดะพะนะดัะผ ะดะพ ะบะพะฝัะฐ
    while a_total - g_total > 500 and i < n:
        a_total -= a_eggs_costs[i]
        result[i] = "G"
        g_total += g_eggs_costs[i]
        i += 1

    # ะัะปะธ ะฝะต ะฒะพะทะผะพะถะฝะพ ัะฐัะฟัะตะดะตะปะธัั
    if a_total - g_total > 500:
        return -1
    else:
        return result


a_costs = []
g_costs = []
in_n = int(input())

for i in range(in_n):
    a, g = list(map(int, input().split()))
    a_costs.append(a)
    g_costs.append(g)

r = uncle_J_task(in_n, a_costs, g_costs)
print(*r)
```
