### Задание №1
Массив A состоит из целых положительных чисел длин отрезков. Составьте из трех отрезков такой треугольник, чтобы его периметр был максимально возможным. Если невозможно составить треугольник с положительной площадью функция возвращает 0.

```python
import random
```

```python
def checkTriangle(a, b, c):
    if (a + b <= c):
        return False
    elif (a + c <= b):
        return False
    elif (b + c <= a):
        return False
    return True

def findTriangle(arr):
    for i in range(len(arr) - 2):
        if (checkTriangle(arr[i], arr[i + 1], arr[i + 2])):
            return [arr[i], arr[i + 1], arr[i + 2]]
    return [0,0,0]

user_len_arr = int(input("Введите длину массива: "))

forged_arr = [random.randint(1, 2000) for i in range(user_len_arr)]

print("Созданный массив: ", forged_arr)

forged_arr = sorted(forged_arr, reverse=True)

found_sides = findTriangle(forged_arr)

print("Максмальный периметр: ", sum(found_sides))
print("a = ", found_sides[0])
print("b = ", found_sides[1])
print("c = ", found_sides[2])
```

### Задание №2
Дан массив неотрицательных целых чисел nums. Расположите их в таком порядке, чтобы вместе они образовали максимально возможное число. Замечание: Результат может быть очень большим числом, поэтому представьте его как string, а не integer.

```python
import random

user_m = int(input())

rand_arr = [str(random.randint(0, 1000))  for i in range(user_m)]
max_number = []
max_int = 0

print(rand_arr)
    
for i in range(len(rand_arr)):
    for j in range(i, len(rand_arr)):
        # Лексикографическое сравнение
        if(int(rand_arr[i] + rand_arr[j]) >= int(rand_arr[j] + rand_arr[i])):
            max_int = rand_arr[i]
        else:
            max_int = rand_arr[j]
            # Делаем так, чтобы найденный элемент больше не просматривался
            rand_arr[i], rand_arr[j] = rand_arr[j], rand_arr[i]
            
    max_number.append(max_int)

print(max_number)
```
