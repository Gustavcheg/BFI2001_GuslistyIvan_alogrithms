# Задание 1

Даны две строки: `s1` и `s2` с одинаковым размером.

Проверьте, может ли некоторая перестановка строки `s1` “победить” некоторую перестановку строки `s2` или наоборот.

Строка x может “победить” строку y (обе имеют размер n), если `x[i] >= y[i]` (в алфавитном порядке) для всех i от 0 до n-1.


## Ввод

`abc`
`xya`

## Вывод

`True`

### Исполняемый код записать в ячейку ниже:

```python
def stringGame(s1, s2):
    sorted_s1 = sorted(s1)
    sorted_s2 = sorted(s2)
    check = False
    
    for i in range(len(s1)):
        if (sorted_s1 >= sorted_s2):
            check = True
        else:
            check = False
            
    for i in range(len(s1)):
        if (sorted_s1 <= sorted_s2):
            check = True
        else:
            check = False
            
    return check

user_str1 = input()
user_str2 = input()
print(stringGame(user_str1, user_str2))
```

# Задание 2

Дана строка `s`, вернуть самую длинную полиндромную подстроку в `s`.

## Ввод

`babad`

## Вывод

`aba` или `bab`

## Исполняемый код записывать в ячейку ниже

```python
user_str = input()
max_str = ""
max_str_len = 0

for i in range(len(user_str)):
    for j in range(len(user_str)):
        substring = user_str[i:j + 1]
        if (len(substring) >= max_str_len and substring == substring[::-1]):
            max_str_len = len(substring)
            max_str = substring
            
print(max_str)
```

# Задание 3

Вернуть количество отдельных непустых подстрок текста, которые могут быть записаны как конкатенация некоторой строки с самой собой (т.е. она может быть записана, как `a + a`, где `a` - некоторая строка).

## Ввод

`aabb`

## Вывод

2

## Исполняемый код записывать в ячейку ниже

```python
user_str = input()
cnt_concat_chars = 0
cnt_same_chars = 0

for i in range(1, len(user_str)):

    if (user_str[i] == user_str[i - 1]):
        cnt_same_chars += 1
    else:
        cnt_concat_chars += cnt_same_chars
        cnt_same_chars = 0 
        
cnt_concat_chars += cnt_same_chars

print(cnt_concat_chars)
```
